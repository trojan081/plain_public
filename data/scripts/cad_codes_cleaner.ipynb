```python
import pandas as pd
import re
```

## 1. Regions texts cleaning

```python
df = pd.read_csv('REGIONS_FINAL_v1.csv')
```

```python
def clean_cad_mtext(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    text = re.sub(r'%%[cC]', 'Ø', text)
    text = re.sub(r'%%[dD]', '°', text)
    text = re.sub(r'%%[pP]', '±', text)
    text = re.sub(r'\\P', ' ', text)
    text = re.sub(r'\\[A-Za-z0-9~|\-]+[^;]*;', '', text)
    text = re.sub(r'\\[LlOoKkX]', '', text)
    text = re.sub(r'[{}]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text


df_cleaned = df.copy()

print("Text cleaning...")
df_cleaned['Text'] = df_cleaned['Text'].apply(clean_cad_mtext)

output_csv = 'REGIONS_FINAL_v2.csv'
df_cleaned.to_csv(output_csv, index=False, encoding='utf-8-sig')

print("---")
print(f"New file saved as: {output_csv}")
print(f"Texts cleaned: {len(df_cleaned)}")
```

## 2. Moscow texts cleaning

```python
msk_df = pd.read_csv('MSK_in_progress_1.10.csv')
```

```python
msk_df['Text'] = msk_df['Text'].apply(clean_cad_mtext)

output_file = 'MSK_in_progress_1.10_TEXT_CLEANED.csv'
msk_df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"Global cleanup complete. Data saved to {output_file}")

samples = ["\H1x;\W0.705882; ПК N62-ЗАО", "\H1x;\W0.705882;ТОННЕЛЬ"]
for s in samples:
    print(f"Before: {s} | After: {clean_cad_mtext(s)}")
```
